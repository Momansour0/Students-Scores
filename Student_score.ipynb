## Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
import seaborn as sns
%matplotlib inline
```

<a id='import'><a/> 
## Importing data

```python
path= "student_scores.csv"
df = pd.read_csv(path)
df.head()
```

<h3>Regression Plot</h3>
<a id='reg'><a/> 


```python
# regplot between Hours and Scores
plt.figure(figsize=(10,5))
ax = sns.regplot(data=df, x="Hours",y="Scores")
plt.show
```

<h3>Linear regression</h3>
<a id='lin'><a/> 


```python
# put Hours in var x and scores in var y
x= df[["Hours"]]
y = df["Scores"]
```

```python
# split our data into traing and testing sets
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0) 
```

```python
# fit the training data
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
# check how close  the data is to the fitted regression line for testing data.
lm.score(x_test, y_test)
```

```python
# check how close  the data is to the fitted regression line for training data.
lm.score(x_train, y_train)
```

- We can see the R^2 is much similar using the test data compared to the training data.
- the data fitted by 95%


## Make predcition
<a id='pred'><a/> 


```python
#predict the testing data
y_pred = lm.predict(x_test)
```

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

```python
# calculate the mean abosolute error to know the diffrence between actual and predicted values 
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred)) 
```

```python
# predict the score value for no.hours equal 9.25
hours = np.array(9.25).reshape(-1, 1)
pred = lm.predict(hours)
print(f"the Score value for 9.25 Hours {pred} ")
```
